```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_1 = np.linspace(-5,10,100)
x_2 = np.linspace(0,15,100)

X_1, X_2 = np.meshgrid(x_1,x_2)

a = 1
b = (5.1/(4*np.square(np.pi)))
c = (5/np.pi)
r = 6
s = 10
t = (1/(8*np.pi))

f = a*(X_2 - b*(X_1**2) + c*X_1 - r)**2 + s*((1 - t)*np.cos(X_1)) + s + 5*X_1

fig, ax = plt.subplots()
cp = ax.contourf(X_1, X_2, f, levels=15, cmap='viridis')
fig.colorbar(cp)

ax.set_xlabel('x1')
ax.set_ylabel('x2')

plt.show()
```
