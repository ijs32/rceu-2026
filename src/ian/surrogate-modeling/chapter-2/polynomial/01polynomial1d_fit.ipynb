```python
import sys
# todo Windows
sys.path.append("../cpcrr")

import numpy as np

from numpy.polynomial import Polynomial


import cpcrr as cp



def polynomial_print(x, y, max_order = 15):
    xplot = np.linspace(0,1, 200)
    for order in range(max_order):
        p = Polynomial.fit(x, y, order)
        # the curve of the fitting polynomial
        yplot = p(xplot)
        cp.functions.quickplot_1d(xplot, yplot, f"ord{order}", x, y)


if __name__ == '__main__':

    f = cp.functions.onevar

    x, y = cp.functions.function_noise_1d(f, xmin = 0, xmax = 1, npoints=20, stddev = 0.5)
    cp.functions.quickplot_scatter(x, y, f"Data set")

    polynomial_print(x, y, max_order=15)


```
